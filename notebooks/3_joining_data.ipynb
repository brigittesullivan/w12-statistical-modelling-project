```python
# import foursquare data for montreal
path = '/Users/brigitteasullivan/My Drive/0.Bootcamp/1. Data Course/lighthouse-data-notes/Week_12/Project/w12-statistical-modelling-project/data/fsq_data.csv'
fsq_data = pd.read_csv(path)
fsq_data 
```

```python
# data cleaning - merge latitude and longitude to match to stations. 


stations_data['location'] = stations_data['latitude'].astype(str) +","+ stations_data['longitude'].astype(str) 
stations_data = stations_data.rename(columns = {'location': 'station_location'})
```

```python
# add a 'green space count' for each station location
fsq_data_outdoors = fsq_data.groupby(['fsq_location']).count().reset_index()
fsq_data_outdoors
```

```python
fsq_data_outdoors = fsq_data_outdoors.drop(columns=['category_name', 'fsq_name', 'distance to station', 'fsq_id'])
```

```python
fsq_data_outdoors = fsq_data_outdoors.rename(columns={'categoryid': 'outdoor_space_num'})
fsq_data_outdoors
```

```python
# also look at number of parks 
fsq_data_outdoors_cat = fsq_data.groupby(['fsq_location', 'category_name']).count().reset_index()
fsq_data_outdoors_cat
```

```python
fsq_data_outdoors_cat = fsq_data_outdoors_cat.drop(columns=['fsq_name', 'distance to station', 'fsq_id'])
```

```python
fsq_data_outdoors_cat = fsq_data_outdoors_cat.rename(columns={'categoryid': 'num_by_cat'})
fsq_data_outdoors_cat
```

```python
# merge.... for every row in stations data, show all the records with a matching location in fsq data

stations_sample = stations_data[:1]
stations_sample

data = pd.merge(stations_data, fsq_data_outdoors, how='inner', left_on = stations_data['station_location'], right_on = fsq_data_outdoors['fsq_location'])
data
```

```python
# perform another merge to bring in the num of outdoor spaces by category
```

```python
data = data.drop(columns=['key_0', 'fsq_location'])
all_data = pd.merge(data, fsq_data_outdoors_cat, how='inner', left_on = data['station_location'], right_on = fsq_data_outdoors_cat['fsq_location'])
all_data
```

## b. import foursquare data

```python
# import foursquare data for montreal
path = '/Users/brigitteasullivan/My Drive/0.Bootcamp/1. Data Course/lighthouse-data-notes/Week_12/Project/w12-statistical-modelling-project/data/fsq_data.csv'
fsq_data = pd.read_csv(path)
fsq_data 
```

## c. data cleaning / wrangling

```python
# data cleaning - merge latitude and longitude to be able to perform merge using one column

stations_data['location'] = stations_data['latitude'].astype(str) +","+ stations_data['longitude'].astype(str) 
stations_data = stations_data.rename(columns = {'location': 'station_location'})
```

```python
#view df
stations_data
```

```python
#cleaning: review the different categories to see if any categories should be removed from data

fsq_data['category_name'].value_counts()

# decision based on results that all categories can be included in analysis. 
```

```python
# print(type())
# QA / testing  
fsq_data[fsq_data['fsq_location'] == stations_data['station_location'][0]]

```

```python
# add a 'green space count' for each station location
fsq_data_outdoors = fsq_data.groupby(['fsq_location']).count().reset_index()
fsq_data_outdoors
```

```python
fsq_data_outdoors = fsq_data_outdoors.drop(columns=['category_name', 'fsq_name', 'distance to station', 'fsq_id'])
```

```python
fsq_data_outdoors = fsq_data_outdoors.rename(columns={'categoryid': 'outdoor_space_num'})
fsq_data_outdoors
```

```python
# also look at number of results by category (parks, monuments, dog parks, etc...)
fsq_data_outdoors_cat = fsq_data.groupby(['fsq_location', 'category_name']).count().reset_index()
fsq_data_outdoors_cat
```

```python
fsq_data_outdoors_cat = fsq_data_outdoors_cat.drop(columns=['fsq_name', 'distance to station', 'fsq_id'])
```

```python
fsq_data_outdoors_cat = fsq_data_outdoors_cat.rename(columns={'categoryid': 'num_by_cat'})
fsq_data_outdoors_cat
```

### MERGE

```python
# merge.... for every row in stations data, show all the records with a matching location in fsq data. 
# QA: first created a sample stations df to confirm merge output is as expected

stations_sample = stations_data[:1]
stations_sample

# perform merge on entire DF's (replaced "stations_sample" with "stations_data"
data = pd.merge(stations_data, fsq_data_outdoors, how='inner', left_on = stations_data['station_location'], right_on = fsq_data_outdoors['fsq_location'])
data
```

```python
# perform another merge to bring in the num of outdoor spaces by category
```

```python
data = data.drop(columns=['key_0', 'fsq_location'])
all_data = pd.merge(data, fsq_data_outdoors_cat, how='inner', left_on = data['station_location'], right_on = fsq_data_outdoors_cat['fsq_location'])
all_data
```

```python
# QA - confirm join occured as expected. Randomly chose index 100. 
test_location = stations_data['station_location'][100]
print(test_location)
                                                  
fsq_data_outdoors_cat[fsq_data_outdoors_cat['fsq_location'] == test_location]
```

```python
# extract one record for testing / QA
stations_data[100:101]
```

```python
stations_data[stations_data['station_location'] == test_location]
```

```python
#check output for merged data
all_data[all_data['station_location'] == test_location]

# total number of outdoor space for this station location is 50... check the sum of num by cat is also 50 
```

```python
# sum of results by category is also 50... joins function as expected 
all_data[all_data['station_location'] == test_location]['num_by_cat'].sum()
```

```python
#add number of parks as a column to df, first create DF with data required station location and num of parks
fsq_data_parks = fsq_data_outdoors_cat[fsq_data_outdoors_cat['category_name'] == 'Park'].rename(columns={'num_by_cat': 'num_parks'}).drop(columns=['category_name']).reset_index(drop=True)
fsq_data_parks
```

```python
all_data
```

```python
# merge num parks with all data 
all_data = all_data.drop(columns=['key_0']) # created in last merge and not required 
all_data = pd.merge(all_data, fsq_data_parks, how='inner', left_on = all_data['station_location'], right_on = fsq_data_parks['fsq_location'])
all_data
```

```python
# QA - in all data location has 10 parks, should also show 10 parks
fsq_data_parks[fsq_data_parks['fsq_location'] == '45.49487159597493,-73.57112377882004']
# output - yes merge worked well
```

```python
# more cleaning... remove duplicate columns - location. 
# review all columns to see which ones can be removed / are duplicated
all_data.columns
```

```python
# decision to remove key, id latitude, long, fsq location since they are all duplicated columns
all_data = all_data.drop(columns=['key_0', 'id', 'latitude', 'longitude', 'fsq_location_x', 'fsq_location_y'])
all_data
```
